```python
#create dataframe with the array
data={'weight':[66,75,84,96,48,71],
      'height':[156,165,186,165,174,167]}
bmi=pd.DataFrame(data)
bmi
```

```python
# create a bmi column. formula: (weight(kg))/(height(m))^2
bmi["BMI"]=bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
bmi.loc[3]=[78,np.nan,np.nan]
bmi
```

```python
#insert another row with missing values
bmi.loc[5]=[np.nan,69,np.nan]
bmi
```

```python
#replace missing value with the mean value of that column
bmi["bmi_new"]=bmi["weight"]/(bmi["height"]/100)**2
bmi
```
